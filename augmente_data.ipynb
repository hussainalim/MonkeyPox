```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
import os

# Create an ImageDataGenerator with additional brightness adjustment
datagen = ImageDataGenerator(
    rotation_range=20,         
    brightness_range=[0.2, 1.4], 
    zoom_range=0.2,           
    horizontal_flip=True,      
    fill_mode='nearest',       
    
)

# Path to the folder containing images to augment
image_folder_path = 'fold1\Train\Chickenpox'
save_dir = 'augmented_Chickenpox'  # Folder where augmented images will be saved

# Create the directory if it doesn't exist
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# Loop through each image in the folder
for img_file in os.listdir(image_folder_path):
    # Construct the full path to the image file
    img_path = os.path.join(image_folder_path, img_file)
    
    # Load the image
    img = image.load_img(img_path,target_size=(224, 224))
    x = image.img_to_array(img)
    x = x.reshape((1,) + x.shape)  # Reshape the image for the ImageDataGenerator
    
    # Generate augmented images and save them
    i = 0
    for batch in datagen.flow(x, batch_size=1, save_to_dir=save_dir, save_prefix='aug_' + img_file.split('.')[0], save_format='jpeg'):
        i += 1
        if i >= 7:  
            break

print(f"Augmented images with brightness variation have been saved in '{save_dir}' directory.")
```
